```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
job_roles = {
    "Data Scientist": "python sql machine learning data analysis statistics pandas numpy",
    "DevOps Engineer": "aws docker kubernetes ci cd automation cloud linux",
    "Backend Developer": "java python sql apis rest databases spring",
    "Frontend Developer": "javascript react html css ui design web design",
    "Cloud Architect": "aws azure cloud computing networking security infrastructure",
    "Machine Learning Engineer": "python machine learning tensorflow pytorch deep learning neural networks",
    "Cybersecurity Analyst": "network security penetration testing firewalls encryption threat analysis",
    "Mobile App Developer": "flutter kotlin swift android ios mobile development",
    "Database Administrator": "sql database management mysql postgresql indexing backup",
    "Data Analyst": "excel sql data visualization tableau power bi reporting statistics",
    "Full Stack Developer": "javascript react node express mongodb html css apis",
    "AI Research Engineer": "python deep learning research pytorch tensorflow nlp computer vision",
    "Network Engineer": "networking cisco routing switching firewalls vpn security",
    "QA Automation Engineer": "selenium testing automation python java test cases quality assurance",
    "Site Reliability Engineer": "linux monitoring automation kubernetes docker incident response cloud",
    "Blockchain Developer": "solidity ethereum blockchain smart contracts web3 cryptography",
    "Game Developer": "unity c sharp game design 3d graphics physics animation",
    "UI UX Designer": "figma design wireframes prototyping user research ui ux",
}
```

```python
user_skills = ["Python", "Cloud Computing", "Automation"]
user_profile = " ".join(user_skills).lower()
```

```python
roles = list(job_roles.keys())
descriptions = list(job_roles.values())
descriptions.append(user_profile)
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(descriptions)
```

```python
user_vector = tfidf_matrix[-1]
role_vectors = tfidf_matrix[:-1]

scores = cosine_similarity(user_vector, role_vectors).flatten()
```

```python
ranked = sorted(zip(roles, scores), key=lambda x: x[1], reverse=True)
top_3 = ranked[:3]

print("Based on your skills:", user_skills)
print("Top 3 recommended career paths:\n")
for rank, (role, score) in enumerate(top_3, start=1):
    print(f"{rank}. {role} — Match Score: {round(score, 2)}")
```
